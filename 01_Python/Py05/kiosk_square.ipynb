#### 추가 주문

* 음료 하나만 주문할 수도 있지만, 주문을 추가로 더 할 수도 있다.   
* 그렇기 때문에 주문을 하고 나서 추가 주문이 있는지 물어보고 주문을 더 받을 준비  

* 손님에게 추가 주문을 할 것인지 물어보고,  
* 추가 주문을 하지 않고 주문 금액을 지불하려면 0을 눌러서 주문이 완료되었다고 알려준다  

* 0이 아닐 동안은 추가 주문을 받을 수 있어야 하니 while n != 0 문으로 추가 주문을 반복

* 주문표에 모든 메뉴를 출력해야 하므로   
주문 리스트(self.order_menu)를 만들어서 주문한 메뉴들을 추가해놓고,  
* 그에 해당하는 가격도 가격 리스트(self.oreder_price)에 추가  
  
* 주문과 가격 리스트는 menu_select() 메소드를 호출할 때 제일 처음에 빈 리스트를 만들고,  
* 손님이 음료를 정하면 가격 리스트를 추가(append())  

> self.order_menu = []  # 주문 리스트  
> self.order_price = []  # 가격 리스트  

> self.order_price.append(self.price[n-1])  # 가격 리스트에 추가  
> self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가  

> print(self.order_menu, self.order_price)
  
  
#### 지불

* 지불(pay) 메서드에는 지불할 금액을 출력하고,  
지불 수단을 현금 또는 카드 중에서 고른다.  

* 현금을 선택하면 직원을 호출하겠습니다.라고 출력하고  
카드를 선택하면 IC칩 방향에 맞게 카드를 꽂아주세요.를 출력하는 메서드  

* 1이나 2 외의 다른 번호를 선택했을 경우에는  
다시 결제를 시도해 주세요.라는 메시지가 나오는 예외 처리  

> 현금('cash' or 1) → '직원을 호출하겠습니다.'  
> 카드('card' or 2) → 'IC칩 방향에 맞게 카드를 꽂아주세요.'  




```python
# [:3] coffee, [3:] noncoffee
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]


class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트

        print()  # 줄 바꿈
        n = 0
        while n < 1 or len(menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일때
            if 1 <= n & n <= len(menu):
                self.price_sum = self.price[n-1]  # 선택 음료의 가격

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        self.order_price.append(self.price[n-1])  # 가격 리스트에 추가
        self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가

        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price_sum, '원')  # 온도 속성을 추가한 주문 결과 출력

        self.add_select()



    # 추가 주문 메서드
    def add_select(self):

      while True : # n == 0 면 break
        print()  # 줄 바꿈
        n = int(input("추가 주문은 1을, 결제는 0을 입력하세요 : "))

        if n == 0:
          break

        elif n == 1:
          print()  # 줄 바꿈
          v = 0
          while v < 1 or len(menu) < v:
            v = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일때
            if 1 <= v & v <= len(menu):
                self.price_sum = self.price[v-1]  # 선택 음료의 가격

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

          t = 0
          while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1과 2 중 하나를 입력하세요.")

        else:
          print("1과 0 중 하나를 입력하세요.")

          self.order_price.append(self.price[v - 1])  # 가격 리스트에 추가
          self.order_menu.append(self.temp + ' ' + self.menu[v - 1])  # 주문 리스트에 추가

          print('추가 주문 음료', self.temp, self.menu[v - 1], ' : ', self.price_sum, '원')



    # 지불 메서드
    def pay(self):

      print()  # 줄 바꿈
      print("결제 페이지")
      print(f'최종 주문 음료: {", ".join(self.order_menu)}') # f-string에서 문자열 리스트 요소만 구분자 넣어서 출력
      print(f'최종 결제 금액 : {sum(self.order_price)}원')
      p = 0
      while p != 1 and p != 2:  # 1이나 2를 입력할 때까지 물어보기
        p= int(input("결제 수단을 선택해 주세요.(1번 현금, 2번 카드): "))

        if p == 1:
            self.payment = "cash"
            print("직원을 호출하겠습니다.")
        elif p == 2:
            self.payment = "card"
            print("IC칩 방향에 맞게 카드를 꽂아주세요.")
        else:
            print("1과 2 중 하나를 입력하세요.")


    # 주문서 출력 메서드
    def table(self):
      print()  # 줄 바꿈
      # 외곽
      print('⟝' + '-' * 30 + '⟞')
      for i in range(5):
          print('|' + ' ' * 31 + '|')

      # 주문 상품명 : 해당 금액
      for i in range(len(self.order_menu)):
          print(f'    {self.order_menu[i]} : {self.order_price[i]}')
      print()
      print( '    ' + '= ' * 12)
      print(f'    합계 금액 : {sum(self.order_price)}')

      # 외곽
      for i in range(5):
          print('|' + ' ' * 31+ '|')
      print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay() # 지불
a.table() # 주문서 출력
```

평행사변형, 사다리꼴 넓이를 구하는 메서드

```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시 : 밑변,높이\n >>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
        lower, upper, height = map(int, input('밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n >>>').split(','))
        area = (lower + upper) * height / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

s1 = Square()
s1.trape()
```
