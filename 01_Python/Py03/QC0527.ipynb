[QUEST 04] 파이썬을 사용하여 간단한 계산기 프로그램 만들기
===========================================

* 사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램을 만들어주세요

* 조건
  * 사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다
  * 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 다시 입력받기를 시도합니다.
  * 나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.
  * 사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.
  * math 모듈을 사용하여 제곱 연산(**)을 지원합니다.
  * 계산기를 통해 계속 계산을 할 것인지 입력받습니다.


```python
from math import pow


# 유효한 정수 확인 함수
def valid_number(input_text):
  # 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력 -> except
  # 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 -> print
  # 다시 입력받기를 시도 -> while True

  while True: # 완료될 때까지 무한반복
    try:
      num = int(input(input_text))

    # 정수가 아닌 경우 오류 메시지 출력
    except ValueError:
      print("잘못된 입력입니다. 정수를 입력해주세요.")
    else:
      return num


# 유효한 연산자 확인 함수
def valid_operator():

  while True: # 완료될 때까지 무한반복
    # 조건으로 지정된 연산자 리스트

    valid_operator = ["+", "-", "*", "/", "**"]
    operator = input("연산자를 입력하세요 (+,-,*,/,**): ")


    # valid_operator에 들어있는 연산자가 operator로 입력되었다면 operator를 반환
    if operator in valid_operator:
      return operator

    # operator로 지정된 연산자가 아닌 입력이 들어올 경우
    else:
      print("지원하지 않는 연산자입니다.")



# 계산 진행 함수
def calculation(num1, num2, operator):

  if operator == "+":
    result = num1 + num2

  elif operator == "-":
    result = num1 - num2

  elif operator == "*":
    result = num1 * num2

  elif operator == "/":
    try:
      result = num1 / num2
    except ZeroDivisionError:
      result = "0으로 나눌 수 없습니다."

  # math 모듈로 import한 제곱 함수 사용
  elif operator == "**":
    result = pow(num1, num2)

  return result

# 계산기 종료 확인 함수
# True를 반환하면 계산기 종료
def choice_end():
  while True:
    choice = input("계산기를 계속 사용하시겠습니까? (y/n): ")

    # 계산기를 종료할지, 재실행할지 선택
    if choice.lower() == "n":
      return True
    elif choice.lower() == "y":
      return False
    else:
      print("잘못된 입력입니다. 다시 입력해주세요.")


# 계산기 메인 함수
def calculator_app():

  print(f"{'= '*10}계산기를 시작합니다.{' ='*10}")

  while True:

    # 정수를 입력 받고 유효성 확인 후 정수 반환
    num1 = valid_number("첫 번째 정수를 입력하세요: ")
    num2 = valid_number("두 번째 정수를 입력하세요: ")

    # 연산자를 입력 받고 유효성 확인 후 연산자 operator 반환
    operator = valid_operator()

    # 두 정수와 연산자를 계산 함수를 호출하여 연산하고 결과값 result 반환
    result = calculation(num1, num2, operator)

    # 반환된 정수, 연산자, 결과값 출력
    print(f"{num1} {operator} {num2} = {result}")

    # 계산기를 통해 계속 계산할 것인지 입력 받고 확인 후 bool 반환
    # choice_end() == True 면 계산기 종료, False면 계산기 재진행
    if choice_end():
      break


  print(f"{'= '*10}계산기를 종료합니다.{' ='*10}")



### 실행
calculator_app()
```

#### 막히거나 수정한 부분
* 처음 틀을 잡고 짤 때 실행 담당(calculator_app) 함수, 정수 유효성(valid_number) 함수, 연산자 유효성(valid_operator) 함수, 계산기 연산(calculation) 함수를 따로 나누지 못하고 정수+연산자 부분 같이 진행하다가 생각보다 더 세세하게 나눠야 진행 가능한다는 걸 확인하고 기능별로 나누면서 늘려갔다.
* 유효성 확인 시 반복되는 구간의 기능 구현-> while Ture:
* 계산기를 통해 계속 계산을 할 것인지 입력받는 부분을 어떻게 구현해야 계속 진행될지 고민하다가 True를 받으면 실행 함수 에서의 while True를 break하도록 설정했다.
